```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

y_1=sns.load_dataset("penguins").head(20)
y_1
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",size=50)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",palette="Accent")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex",palette="Accent",markers=["o",">"],dashes=True)
plt.show()
plt.grid()
plt.title("python")
```
