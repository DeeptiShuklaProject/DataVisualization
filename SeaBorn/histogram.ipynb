```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

```

```python
var=sns.load_dataset("penguins")
var
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240])
plt.show()
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde="True")
plt.show()
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],rug="True")
plt.show()
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde="True",rug = True,color="r")
plt.show()
```

```python
sns.displot(var["flipper_length_mm"],kde="True",rug = True,color="r",log_scale=True)
plt.show()
```
