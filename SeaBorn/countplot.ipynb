```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.countplot(x="sex",data=var)
plt.show()
```

```python
sns.barplot(x="sex",y="size",data=var)
plt.show()
```

```python
sns.countplot(x="sex",data=var,hue="smoker")
plt.show()
```

```python
sns.countplot(y="sex",data=var,hue="smoker")
plt.show()
```

```python
sns.countplot(x="sex",data=var,hue="smoker",palette="Dark2")
plt.show()
```

```python
sns.countplot(x="sex",data=var,hue="smoker",palette="Dark2",saturation=0.1)
plt.show()
```
