```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",color="g",order=["Fri","Sun","Thur","Sat"],showmeans=True)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,color="r",order=["Fri","Sun","Thur","Sat"],showmeans=True)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,order=["Fri","Sun","Thur","Sat"],showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3,palette="plasma")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=var,showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3,palette="plasma",orient="h")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y=var["total_bill"])
plt.show()
```
