```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var,hue="sex")
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex")
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],palette="BuGn")
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"])
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],y_vars=["total_bill","tip"])
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],kind="reg")
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],kind="kde")
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],kind="scatter")
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],kind="kde",diag_kind="hist")  #scatter,kde,hist,reg
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],markers=["*","<"])
plt.show()
```
