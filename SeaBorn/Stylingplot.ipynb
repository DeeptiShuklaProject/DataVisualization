```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.set_style("darkgrid")
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set_style("dark")
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set_style("white")
sns.barplot(x="day",y="total_bill",data=var)
sns.despine()
plt.show()
```

```python
sns.set_style("white")
plt.figure(figsize=(12,5))
sns.barplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r")
sns.despine()
plt.show()
```

```python
sns.set_style("white")
sns.set_context("poster",font_scale=1)
sns.barplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r")

plt.show()
```

```python
sns.set_style("white")
sns.set_context("paper",font_scale=1)
sns.barplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r")

plt.show()
```

```python
sns.set_style("white")
sns.set_context("talk",font_scale=1)
sns.barplot(x="day",y="total_bill",data=var,palette='cool',hue="sex")

plt.show()
```
