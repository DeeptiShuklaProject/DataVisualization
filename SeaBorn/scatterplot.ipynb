```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=sns.load_dataset("penguins")
var
```

```python
sns.scatterplot(x=var.bill_length_mm,y=var.bill_depth_mm,data=var)
plt.show()
```

```python
sns.scatterplot(x=var.bill_length_mm,y=var.bill_depth_mm,data=var,hue="sex")
plt.show()
```

```python
var=sns.load_dataset("penguins").head(20)
var
```

```python
sns.scatterplot(x=var.bill_length_mm,y=var.bill_depth_mm,data=var,hue="sex",style="sex")
plt.show()
```

```python
sns.scatterplot(x=var.bill_length_mm,y=var.bill_depth_mm,data=var,hue="sex",size="sex",sizes=(100,40))
plt.show()
```

```python
sns.scatterplot(x=var.bill_length_mm,y=var.bill_depth_mm,data=var,hue="sex",size="sex",sizes=(100,40),palette="gist_ncar",alpha=0.5)
plt.show()
```

```python
m={"Male":"*","Female":"o"}
sns.scatterplot(x=var.bill_length_mm,y=var.bill_depth_mm,data=var,hue="sex",style="sex",size="sex",sizes=(100,60),markers=m)
plt.show()
```
