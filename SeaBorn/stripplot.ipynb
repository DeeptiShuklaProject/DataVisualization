```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.stripplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1,edgecolor="r")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1,edgecolor="r",jitter=10)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1,edgecolor="r",size=5,alpha=0.2)
plt.show()
```

```python
n={"Male":"o","Female":"<"}
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1,edgecolor="r",size=5,marker="<")
plt.show()
```
