```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=2)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,palette="winter_r")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var,order=["Dinner","Lunch"])
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=2,color="y")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True)
plt.show()
```

```python
sns.violinplot(x="total_bill",y="day",data=var,hue="time")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var,order=["Dinner","Lunch"],inner="quart") #("box","quartile","point","stick",)
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var,order=["Dinner","Lunch"],inner="box") #("box","quartile","point","stick",)
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var,order=["Dinner","Lunch"],inner="stick") #("box","quartile","point","stick",)
plt.show()
```

```python
sns.violinplot(x=var["total_bill"])
plt.show()
```

```python
sns.violinplot(y=var["total_bill"])
plt.show()
```
